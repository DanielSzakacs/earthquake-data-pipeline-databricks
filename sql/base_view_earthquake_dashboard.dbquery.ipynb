```sql
CREATE OR REPLACE VIEW gold_earthquake_dashboard_vw AS
SELECT
  id,
  longitude,
  latitude,
  elevation,
  title,
  place_description,
  sig,
  mag,
  magType,
  time AS event_time,
  updated AS updated_time,
  country_code,
  sig_class,
  CASE
    WHEN mag < 2 THEN 'Micro (<2.0)'
    WHEN mag < 4 THEN 'Minor (2.0-3.9)'
    WHEN mag < 5 THEN 'Light (4.0-4.9)'
    WHEN mag < 6 THEN 'Moderate (5.0-5.9)'
    WHEN mag < 7 THEN 'Strong (6.0-6.9)'
    ELSE 'Major (7.0+)'
  END AS magnitude_category
FROM gold
WHERE latitude IS NOT NULL
  AND longitude IS NOT NULL
  AND mag IS NOT NULL;
```
