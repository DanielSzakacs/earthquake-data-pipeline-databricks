```sql
SELECT
  magnitude_category,
  COUNT(*) AS earthquake_count
FROM gold_earthquake_dashboard_vw
GROUP BY magnitude_category
ORDER BY
  CASE magnitude_category
    WHEN 'Micro (<2.0)' THEN 1
    WHEN 'Minor (2.0-3.9)' THEN 2
    WHEN 'Light (4.0-4.9)' THEN 3
    WHEN 'Moderate (5.0-5.9)' THEN 4
    WHEN 'Strong (6.0-6.9)' THEN 5
    WHEN 'Major (7.0+)' THEN 6
  END;
```
